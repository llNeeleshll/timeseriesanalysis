Labelling the Plot

```python
title = 'Title'
xlabel = "X Label"
ylabel = 'Y Label'
```

```python
ax = df['Close'].plot(figsize=(12,6), title=title)
ax.autoscale(axis='x', tight=True) # this will make the max of x axis in plot value as the max value to x and so as for y. Both will make it for both
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
ax = df['Close'].plot(figsize=(12,6), title=title, xlim=['2017-01-01','2017-12-31'])
```

Changing linestyle

```python
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,6), title=title, ls='--')
```

Changing line color

```python
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12,6), title=title, ls='--', c='red')
```

Formatting the Date axis

```python
from matplotlib import dates

ax = df['Close'].plot(figsize=(14,5), title=title, xlim=['2017-01-01','2017-03-01'], ylim=[50,60])

# Let's remove the x label first to make the plot cleaner
ax.set(xlabel='');

# Let's now format the xaxis
ax.xaxis.set_major_formatter(dates.DateFormatter('\n%d'))

# Let's set the minor axis now
ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%b'))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```

Monthly and Quarterly plots

```python
df1 = pd.read_csv(r'C:\Users\nevishwakarma\Documents\Datasets\Data_Udemy\airline_passengers.csv', index_col=['Month'], parse_dates=True)
df1.index.freq = 'MS'
df1.head()
```

```python
from statsmodels.graphics.tsaplots import month_plot, quarter_plot
```

```python
month_plot(df1['Thousands of Passengers']);
```

```python
# Resmapling Montly data to quartly to plot quarterly plot.
df_quarter = df1['Thousands of Passengers'].resample('Q').mean()
```

```python
quarter_plot(df_quarter);
```
