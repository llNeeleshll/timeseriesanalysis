**Scatter Plots**

```python
df1.plot.scatter(x='A',y='B', figsize=(10,10))
```

Visualizing 3-dimensions of the data using scatter plots

```python
# Color is by the value in column C. So Color is represnting one more dimension.

df1.plot.scatter(x='A',y='B', c='C', figsize=(15,10), cmap='coolwarm')
```

```python
# Size is by the value in column C. So Color is represnting one more dimension.

df1.plot.scatter(x='A',y='B', s=df1['C'])
```

Representing 4-dimensions of data, using color and size both

```python
df1.plot.scatter(x='A',y='B', c='C', s=df1['D'] * 50, figsize=(15,10), cmap='coolwarm')
```
